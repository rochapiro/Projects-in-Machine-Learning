## Project 1: World Happiness Data

Project description: 
This project consisted in using the World happiness data set, attached in the data folder for that particular project, to fit a model that can best predict the happiness score for a particular country using country-level varaibles

Data set description: 
 The data set consisted of observations at the country level, that is, each row in the data represents a country; while each column represents a particular score for that country. The data set contains variables such as GDP per capita, healthy life expectations, and freedom to make life choices. The countries were ranked into 5 different happiness groups; average, low, high, very low, and very high. 

Key findings from the exercise:  
We can see that the model with the highest average f1 score was the Decision Tree Classifier model, while the one with the lowest average score was the Kneighbors classifier. While the models with the highest standard deviation for the f1 scores were sequential type models.

## Project 2: COVID Image detection



 Project description: 
This project consisted in using x-ray images of lungs that had COVID, pneumonia, or nothing and attempted to classify new X-ray images as well as possible into one of these 3 categories

Data set description: 
The data was composed of x-ray images from various patients who had suffered from COVID or pneumonia as well as some individuals who had healthy lungs. The data was structured as purely image data which had to be preprocessed to be able to feed it into different model architectures. 

Key findings from the exercise:  
Some of the most relevant findings from this image detection exercise were the fact that proven architectures when it comes to image detection tasks remain on top for a reason, as these were able to extract meaningful information from the data to be able to feed it into. Transfer learning techniques using these aforementioned architectures were also a good method of extracting meaningful information from the data.

## Project 3: Text Classification

 Project description: 
This project consisted in using movie reviews that were classified as positive or negative to develop a model that classified reviews as positive or negative as well. 

Data set description: 
As mentioned above, the data set consisted of movie reviews that individuals gave. These were all classified as positive or negative depending on the content of the review and its reference to the movie. These varied in length and vocabulary as well as the underlying movie being reviewed.  

Key findings from the exercise:  
Some of the most relevant findings from this text classification exercise were the following. There was not a positive relationship between model depth and f1 score as we can see in the graph below, that the best fitting line in the scatter plot has a negative slope implying that an increase in depth, on average, was associated with a decrease in the model's f1 score. Also, the number of CONV1D layers does not affect the best performance of the model, however, the models with only 1 layer had the most condensed spread for f1 scores, which were also relatively high. The models that used 3 layers showed the largest spread in scores and the lowest median f1 score.
